```python
import random
from functools import reduce

print("Laboratorio 2 Programacion Cientifica\n")
print("Actividad 1 – Uso de lambda")
#Actividad 1 – Uso de lambda
funcion = lambda x: x**2+2*x+1
print(f"Función evaluada en 2: {funcion(2)}")
print(f"Función evaluada en 4: {funcion(4)}")
print(f"Función evaluada en 6: {funcion(6)}")
print("""Este código crea una función lambda que calcula x**2 + 2*x + 1 y luego imprime el resultado al evaluarla con los valores 2, 4 y 6,
mostrando 9, 25 y 49.\n""")

print("Actividad 2 – Transformación con map()")
#Actividad 2 – Transformación con map()
velocidades = [10, 20, 30, 40, 50]
metros_segundos = list(map(lambda x: round(x/3.6,2), velocidades))
print(f"velocidades k/m: {velocidades}")
print(f"velocidades m/s: {metros_segundos}")
print("""Este código usa map() con una función lambda para convertir una lista de velocidades de km/h a m/s, dividiendo cada valor por 3.6,
y luego imprime la nueva lista con las velocidades convertidas.\n""")

print("Actividad 3 – Filtrado con filter()")
#Actividad 3 – Filtrado con filter()
valores = [2,5,8,11,14,17,20] 
datos_validos = list(filter(lambda x: x%2==0, valores))
print(f"Datos filtrados: {datos_validos}")
print("""Este código usa filter() con una función lambda para filtrar solo los números pares de la lista valores
y luego imprime la nueva lista con los valores [2, 8, 14, 20].\n""")

print("Actividad 4 – Agregación con reduce()")
#Actividad 4 – Agregación con reduce()
numeros = [2,3,4,5]
resultado = reduce(lambda x,y:x+y, numeros)
print(f"Resultado: {resultado}")
print("""Este código usa reduce() con una función lambda para sumar todos los números de la lista [2,3,4,5],
acumulando los valores hasta obtener el resultado final 14.\n""")

print("Actividad 5 – Pipeline de Procesamiento")
#Actividad 5 – Pipeline de Procesamiento
dataset = [12,15,22,35,41,8,3,55,62,90]
resultado = reduce(lambda x,y: x+y,map(lambda x: x*2,
                    filter(lambda x: x>=10, dataset)))
print(f"Resultado: {resultado}")
print("""Este código crea un pipeline de procesamiento donde primero filtra los números mayores o iguales a 10,
luego multiplica cada valor por 2 con map(), y finalmente suma todos los resultados usando reduce(), mostrando la suma total.\n""")

print("Actividad 6 – Simulación de Big Data")
#Actividad 6 – Simulación de Big Data
data = [random.randint(0,100) for _ in range(1000000)]
procesados = list(map(lambda x: x**2, filter(lambda x: x > 50, data)))
promedio = sum(procesados) / len(procesados)
print(f"Promedio: {promedio}")

print("""Este código simula el procesamiento de Big Data generando 1.000.000 de números aleatorios entre 0 y 100,
luego filtra los valores mayores a 50, eleva cada número al cuadrado usando map(), y finalmente calcula el promedio de los valores procesados.\n""")

print("Desafío Avanzado")
#Desafío Avanzado
sensores = [random.uniform(10, 80) for _ in range(500000)]
procesados = list(map(lambda x: x + 273.15, 
                      filter(lambda x: 10 <= x <= 70, sensores)))
promedio_k = sum(procesados) / len(procesados)
print(f"Sensores procesados: {len(procesados)}")
print(f"Temperatura promedio: {promedio_k:.2f} K")
print("""Este código simula sensores industriales generando 500.000 temperaturas aleatorias, luego filtra las que están entre 10 y 70 grados Celcius,
las convierte a Kelvin sumando 273.15, y finalmente calcula y muestra la cantidad de sensores procesados y la temperatura promedio en Kelvin.\n""")

print("""Conclusión:
-Eficiencia en el Código: El uso de map, filter y reduce permite procesar listas grandes de forma más rápida y con mucho menos código que usando bucles for tradicionales.

-Limpieza de Datos: Aplicar filtros (filter) es esencial en ingeniería para eliminar lecturas falsas o errores de los sensores antes de hacer cualquier cálculo matemático.

-Transformación Masiva: La función map facilita la conversión de unidades (como de Celsius a Kelvin o de km/h a m/s) de miles de datos al mismo tiempo sin errores manuales.

-Automatización: El diseño de "pipelines" (conectar un filtro con un mapa y luego una suma) permite automatizar el análisis de datos industriales de manera organizada.""")
 
```
